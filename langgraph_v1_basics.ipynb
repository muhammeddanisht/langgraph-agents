```python
# importing our State dictionary tool
from typing import TypedDict

# importing LangGraph blueprint tools
from langgraph.graph import StateGraph, START, END
```

```python
# State = our dictionary. Robot carries this while working.
# TypedDict = tool that defines what pockets the dictionary has.

class State(TypedDict):
    user_question: str  # pocket 1 = stores the user's question (text)
```

```python
def greet_user(state):
    question = state["user_question"]
    return {"user_question": "Hello! You asked: " + question}
```

```python
# StateGraph = blueprint of the whole factory
graph = StateGraph(State)                        # create blueprint using our State dictionary

# add_node = add worker to blueprint
# first argument = worker's name (text)
# second argument = actual worker function
graph.add_node("greet_user", greet_user)         # add greet_user worker to blueprint

# add_edge = road between workers
graph.add_edge(START, "greet_user")              # factory starts at greet_user
graph.add_edge("greet_user", END)                # factory stops after greet_user

# compile = make blueprint ready to run
app = graph.compile()                            # factory is built. ready to start.
```

```python
# invoke = start the factory. give it the starting dictionary.
result = app.invoke({"user_question": "Book ticket to talkies"})  # give robot its first question

print(result)  # print what robot returned
```

```python
from typing import TypedDict
from langgraph.graph import StateGraph, START, END

# Step 1 — define State dictionary
class State(TypedDict):
    user_question: str              # one pocket — stores text

# Step 2 — worker 1
def greet_user(state):
    question = state["user_question"]                          # read pocket
    return {"user_question": "Hello! You asked: " + question}  # update pocket

# Step 3 — worker 2
def shout(state):
    question = state["user_question"]                          # read pocket
    return {"user_question": question.upper()}                 # make ALL CAPS

# Step 4 — build blueprint
graph = StateGraph(State)                    # create blueprint

graph.add_node("greet_user", greet_user)    # add worker 1
graph.add_node("shout", shout)              # add worker 2

graph.add_edge(START, "greet_user")         # start at worker 1
graph.add_edge("greet_user", "shout")       # worker 1  worker 2
graph.add_edge("shout", END)                # worker 2  stop

app = graph.compile()                       # factory ready

# Step 5 — run it
result = app.invoke({"user_question": "Book ticket to talkies"})
print(result)
```
